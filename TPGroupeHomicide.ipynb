```python
dfcrime = pd.read_csv (r"/Users/abi/Desktop/ipssi/M2/Open_Data_S1/TPGroupe/crimes.csv", index_col=['Libellé index'])
dfcrime
```

```python
newDF = dfcrime.T

newDF = newDF.reset_index()
newDF
```

```python
newDF['Annee'] = newDF['index'].map(lambda x: x.split('_')[0])
newDF
```

```python
DF2Homicide = newDF.groupby(['Annee'])[list(newDF.filter(regex='Homicide'))].sum()
DF2Homicide = DF2Homicide[:-1]
DF2Homicide
```

```python
# ⇧ Nombre des différentes formes d'homicide par année
```

```python
DF2Homicide.plot_bokeh(kind='bar', stacked=True)
```

```python
DF2Tentative = newDF.groupby(['Annee'])[list(newDF.filter(regex='Tentative'))].sum()
DF2Tentative = DF2Tentative[:-1]
DF2Tentative
```

```python
DF2Tentative.plot_bokeh(kind='bar', stacked=True, title='Tentative d\'homicides')
```
