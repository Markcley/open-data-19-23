```python
import pandas as pd

df_2021 = pd.read_excel('./data/crimes-et-delits-enregistres-par-les-services-de-gendarmerie-et-de-police-depuis-2012.xlsx', sheet_name='Services PN 2021')
df_2020 = pd.read_excel('./data/crimes-et-delits-enregistres-par-les-services-de-gendarmerie-et-de-police-depuis-2012.xlsx', sheet_name='Services PN 2020')
df_2019 = pd.read_excel('./data/crimes-et-delits-enregistres-par-les-services-de-gendarmerie-et-de-police-depuis-2012.xlsx', sheet_name='Services PN 2019')
df_2018 = pd.read_excel('./data/crimes-et-delits-enregistres-par-les-services-de-gendarmerie-et-de-police-depuis-2012.xlsx', sheet_name='Services PN 2018')
df_2017 = pd.read_excel('./data/crimes-et-delits-enregistres-par-les-services-de-gendarmerie-et-de-police-depuis-2012.xlsx', sheet_name='Services PN 2017')
df_2016 = pd.read_excel('./data/crimes-et-delits-enregistres-par-les-services-de-gendarmerie-et-de-police-depuis-2012.xlsx', sheet_name='Services PN 2016')
df_2015 = pd.read_excel('./data/crimes-et-delits-enregistres-par-les-services-de-gendarmerie-et-de-police-depuis-2012.xlsx', sheet_name='Services PN 2015')

df_2021.head(5)
```

```python
df_2021_filtered = df_2021.set_index('Départements').drop('Année 2021 - services de police', axis=1).drop('Périmètres', axis=0)

df_2021_filtered.head(5)
```

```python
df_2021_numbers = df_2021_filtered.sum(axis=1).reset_index(name='numbers').set_index('Départements')

df_2021_numbers
```

```python


df_2021_filter = df_2021_numbers.sort_values(by='numbers', ascending=False).head(10)

```

```python
df_2021_filter.plot.pie(y='numbers', autopct='%1.1f%%').legend(bbox_to_anchor=(4.0, 1.0))
```

```python
df_2021_filter = df_2021_filter.T
df_2021_filter.plot(kind='bar').legend(bbox_to_anchor=(1.0, 1.0))
```
