```python
import pandas as pd
import matplotlib.pyplot as plt
crimes_delits = pd.read_excel(open('./xlsx/crimes-delit.xlsx', 'rb'), sheet_name='France_Entière')
pd.options.plotting.backend = "plotly"
```

```python
# crimes et délis sur les 20 dernieres années

df = crimes_delits.drop(['Index', 'libellé index'], axis=1)
df = df.sum(axis=0, numeric_only=True)
df.index = pd.to_datetime(df.index, format='_%Y_%m')
s = df.reindex(index=df.index[::-1])
df = pd.DataFrame(s).reset_index()
df.columns = ['Date', 'crimes et délits']
df.set_index('Date').plot(title="Crimes et délits durant ces 20 dernières années")
```

```python
tab = []

xl = pd.ExcelFile('./xlsx/crimes-delit.xlsx')
for sheet_name in xl.sheet_names:
    try:
        int(sheet_name)
    except(ValueError):
        continue
    xl = pd.read_excel(open('./xlsx/crimes-delit.xlsx', 'rb'), sheet_name=sheet_name)
    xl = xl.drop(['Index', 'libellé index'], axis=1)
    df = xl.sum(axis = 0, numeric_only=True).reset_index(name=sheet_name).set_index('index').transpose()
    tab.append(df)
df = pd.concat(tab)
df.columns = pd.to_datetime(df.columns, format='_%Y_%m')
df.T.plot()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

xl = pd.read_excel(open('./xlsx/crimes-delit.xlsx', 'rb'), sheet_name='France_Entière')
tab = []

filters = {
    'infractions' : 'Infractions', 
    'Proxénétisme|Viols|sexuelles|sexuels': 'Sexuels', 
    'Faux|Fausse|Contrefaçons|Falsification': 'Contrefaçons',
    'vols|cambriolages': 'Vols et cambriolages',
    'coups et blessures': 'Coups et blessures'
}

for key, value in filters.items():
    stat = xl.loc[xl['libellé index'].str.lower().str.contains(key.lower())]
    stat = stat.drop(['Index', 'libellé index'], axis=1)
    stat = stat.sum(axis=0, numeric_only=True)
    stat.index = pd.to_datetime(stat.index, format='_%Y_%m')
    s = stat.reindex(index=stat.index[::-1])
    df = pd.DataFrame(s).reset_index()
    df.columns = ['Date', value]
    df = df.set_index('Date')
    tab.append(df.T)

df = pd.concat(tab)
df.T.plot(title="Crimes et délits regrouper par type ces 20 dernières années")
```
